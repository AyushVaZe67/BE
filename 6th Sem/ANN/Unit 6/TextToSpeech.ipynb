```python
from gtts import gTTS
from IPython.display import Audio

text = "My name is Ayush and my friend's name is Omkar. We are in the classroom."
text1 = "मेरा नाम आयुष है और मेरे दोस्त का नाम ओंकार है। हम क्लासरूम में हैं।"
tts = gTTS(text=text, lang='hi')
tts1 = gTTS(text=text1, lang='hi')
tts.save("output.mp3")
tts1.save("output1.mp3")

Audio("output.mp3")
```
